```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_cities_in_Vietnam"

# Fetch the content of the page
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Locate the table (you may need to refine the class/id selection based on the structure)
tables = soup.findAll('table', {'class': 'wikitable'})

# Extract headers
headers = ['Name']

# Prepare data for CSV
data = []

# Add headers to the data list
data.append(headers)


for table in tables:
    
    # Extract rows
    rows = table.find_all('tr')
    
    # Extract row data
    for row in rows[1:]:  # Skip the header row
        # print(row)
        cols = row.find_all('td')
        data.append([cols[0].text.strip()])
        

# Write data to CSV
with open('provinces_vietnam.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("Data has been written to provinces_vietnam.csv")
```
