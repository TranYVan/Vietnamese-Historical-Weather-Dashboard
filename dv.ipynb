```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import unidecode
import csv

def normalize_text(text):
    if isinstance(text, str):
        text = unidecode.unidecode(text)  # Remove diacritics
        text = text.replace(" ", "").lower()  # Remove spaces and convert to lowercase
    return text


weather = pd.read_csv('./weather.csv', header=0)
province  = pd.read_csv('provinces_vietnam.csv', header=0)
# print(weather['province'].unique())
# print(province['Name'].unique())

weather['province'] = weather['province'].replace('Play Cu', 'Pleiku')
weather['province'] = weather['province'].replace('Qui Nhon', 'Quy Nhon')
weather['province'] = weather['province'].str.replace(' City', '', regex=False)

weather['province_normalized'] = weather['province'].apply(normalize_text)
province['name_normalized'] = province['Name'].apply(normalize_text)

df = pd.merge(weather, province, left_on='province_normalized', right_on='name_normalized', how='left')

# print(weather['province_normalized'].unique())

unmatched_in_df1 = weather[~weather['province_normalized'].isin(df['province_normalized'])]
# print("Rows in df1 with no match in df2:")
unmatched_in_df1

# print(weather)
# print(df)
df.to_csv('data.csv', index=False, encoding='utf-8')
```
